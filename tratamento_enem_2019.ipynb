**CLEAR DATA**

```python
# drop() is to clear columns
#good pratic to create a new variable to manipulation
data1 = data.drop(columns=['CO_MUNICIPIO_RESIDENCIA', 'SG_UF_RESIDENCIA', 'TP_ANO_CONCLUIU'],)
```

```python
data1.head()
```

```python
data1.shape
```

```python
#fix error from grade/nota
data1.loc[:,'NU_NOTA_CN'] /= 10
data1.loc[:,'NU_NOTA_LC'] /= 10
data1.loc[:,'NU_NOTA_MT'] /= 10
data1.loc[:,'NU_NOTA_CH'] /= 10
```

```python
#rename grade / column - clear view
data1 = data1.rename(columns={'NU_NOTA_CN': 'NOTA_CN','NU_NOTA_LC': 'NOTA_LC', 
                              'NU_NOTA_MT': 'NOTA_MT', 'NU_NOTA_CH': 'NOTA_CH'})
```

```python
data1 = data1.rename(columns={'NU_IDADE':'IDADE', 'NU_SEXO':'SEXO', 'TP_COR_RACA':'RACA', 
                              'Q025':'INTERNET', 'TP_ESCOLA':'ESCOLA',})
```

```python
data1 = data1.rename(columns={'NU_NOTA_COMP1': 'NOTA_COMP1','NU_NOTA_COMP2': 'NOTA_COMP2', 
                              'NU_NOTA_COMP3': 'NOTA_COMP3', 'NU_NOTA_COMP4': 'NOTA_COMP4',
                              'NU_NOTA_COMP5': 'NOTA_COMP5'})
```

```python
data1.head()
```

```python
#rename race / row - clear view
#o que vai ser alterado
data1["TP_LINGUA"] = data1["TP_LINGUA"].replace({0:"Ingles", 1:"Espanhol"})
```

```python
data1['IDADE'].value_counts()
```

```python
#contou a menor e maior idade que tentou ENEM e descobriu um erro com 2 anos, por exemplo
data1['IDADE'].value_counts().sort_index()
```

```python
#query to compare column
menores_12 = data1.query('IDADE < 12') ['NO_MUNICIPIO_RESIDENCIA'].value_counts()
```

```python
menores_12
```

```python
#loc : localize
dados_maiores_11 = data1.loc[data1.IDADE > 11]
```

```python
dados_maiores_11.head()
```

```python
#na nova tabela de dados procure em idade, cont e organize
dados_maiores_11['IDADE'].value_counts().sort_index()
```

```python
#treineiros - NEW TABLE
dados_maiores_11.query('IN_TREINEIRO == 1')['IN_TREINEIRO'].value_counts()
```

```python
#treineiros - NEW TABLE
#==1 is ID in dictionary

treineiros = dados_maiores_11.loc[dados_maiores_11.IN_TREINEIRO == 1]
```
